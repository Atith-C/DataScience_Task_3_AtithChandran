```python
# Importing required libraries
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Generating sample data
x = np.linspace(-5, 5, 100)  # Here I Generated 100 points between -5 and 5 for x-axis
y = np.linspace(-5, 5, 100)  # And here I Generated 100 points between -5 and 5 for y-axis
x, y = np.meshgrid(x, y)  # Created a 2D grid of x, y values

# Defining a function for the z-values (e.g., a mathematical surface function)
z = np.sin(np.sqrt(x**2 + y**2))  # z is the sine of the Euclidean distance
```

```python
# Creating a 3D surface plot
fig = plt.figure(figsize=(10, 7))  # Here I Created a figure of size 10x7 inches
ax = fig.add_subplot(111, projection='3d')  # Created a 3D axis

# Plotting the surface
surf = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')  # Used 'viridis' colormap

# Adding labels
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

# Adding a color bar to show the colormap
fig.colorbar(surf)

# To Show the plot
plt.show()
```

```python
# Creating a 3D surface plot with a different colormap
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Used a different colormap 'coolwarm'
surf = ax.plot_surface(x, y, z, cmap='coolwarm', edgecolor='none')

# Adding labels
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

# Adding a color bar
fig.colorbar(surf)

# To Show the plot
plt.show()
```

```python
# Creating a 3D surface plot with shading using LightSource for better lighting
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Shading is applied using the lightsource parameter in plot_surface
# for Matplotlib versions before 3.7
from matplotlib.colors import LightSource

# Create a light source object
ls = LightSource(azdeg=0, altdeg=65)

# Applying the LightSource to the surface plot for shading
surf = ax.plot_surface(x, y, z, cmap='plasma',
                       edgecolor='none', lightsource=ls)  # Applying lightsource

# Adding labels for the axes
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

# Adding a color bar to indicate the colormap values
fig.colorbar(surf)

# To Display the plot
plt.show()
```

```python
# Save the plot as an image
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
fig.colorbar(surf)

# Saving the figure as '3d_surface_plot.png'
plt.savefig('3d_surface_plot.png')

# Show the plot
plt.show()
```
